```python
#importing file
fig = plt.figure(figsize= (16,4))
for i in range(1,4):
    path = f'{i}.txt'
    data = pd.read_csv(path, delim_whitespace=True)
    time = data.iloc[:, 0]
    current = data.iloc[:, 1]
    potential = data.iloc[:, 2]
    scan = data.iloc[:, 3]
    ax = fig.add_subplot(1,3,i)
    ax.scatter(potential,current,s=1, label=f'azobenzene {i}')
    ax.grid(True)
    ax.legend()
```
